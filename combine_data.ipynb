```python
import pandas as pd
```

```python
fat_muscle_info = pd.read_csv('/Users/zilianghong/Documents/GitHub/IPMN-Tissue-Analysis/tissue_from_t1.csv')
fat_muscle_info['name'] = fat_muscle_info['name'].apply(lambda x: str.upper(x.split('.nii.gz')[0]))

panc_cyst_info = pd.read_csv('/Users/zilianghong/Documents/GitHub/IPMN-Tissue-Analysis/tissue_from_t2.csv')
panc_cyst_info['name'] = panc_cyst_info['name'].apply(lambda x: str.upper(x.split('.nii.gz')[0]))
```

```python
panc_cyst_info['name'] = panc_cyst_info['name'].str.replace('AHN_', 'AHN', regex=False)
panc_cyst_info['name'] = panc_cyst_info['name'].str.replace('IUC', 'IU', regex=False)
```

```python
merged_df = pd.merge(panc_cyst_info, fat_muscle_info, on='name', how='inner')
merged_df.sort_values(by='name')
```

```python
label = pd.read_excel('/Users/zilianghong/Documents/GitHub/IPMN-Tissue-Analysis/IPMN_labels_t1.xlsx')
label = label.rename(columns={'Unnamed: 0': 'name','risk assessment':'label'})
label['name'] = label['name'].apply(lambda x: str.upper(x.split('.nii.gz')[0]))
label = label.drop(columns=['pathology labels','follow-up information','Unnamed: 5','Unnamed: 4'])
```

```python
center_list = ['IU','EMC','CAD','NU','MCA','NYU','AHN','NORTHWESTERN','MCF']
for c in center_list:
    merged_df.loc[merged_df['name'].str.contains(c), 'center'] = c
```

```python
merged_df = pd.merge(merged_df,label, on='name', how='outer')
pd.set_option('display.max_rows', None)
merged_df.sort_values(by='name')
```

```python
merged_df = merged_df.replace('NORTHWESTERN', 'NU')
```

```python
merged_df.dropna().to_csv('/Users/zilianghong/Documents/GitHub/IPMN-Tissue-Analysis/IPMN_tissue_data.csv',index=False)
```
