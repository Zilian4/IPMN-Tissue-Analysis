```python
import pandas as pd
```

```python
fat_muscle_info = pd.read_csv('/Users/zilianghong/Documents/GitHub/IPMN-Tissue-Analysis/tissue_from_t1.csv')
fat_muscle_info['name'] = fat_muscle_info['name'].apply(lambda x: str.upper(x.split('.nii.gz')[0]))

panc_cyst_info = pd.read_csv('/Users/zilianghong/Documents/GitHub/IPMN-Tissue-Analysis/tissue_from_t2.csv')
panc_cyst_info['name'] = panc_cyst_info['name'].apply(lambda x: str.upper(x.split('.nii.gz')[0]))
```

```python
panc_cyst_info['name'] = panc_cyst_info['name'].str.replace('AHN_', 'AHN', regex=False)
panc_cyst_info['name'] = panc_cyst_info['name'].str.replace('IUC', 'IU', regex=False)
```

```python
merged_df = pd.merge(panc_cyst_info, fat_muscle_info, on='name', how='inner')
pd.set_option('display.max_rows', None) 
merged_df.sort_values(by='name')
```

```python
merged_df.to_csv('./IPMN_tissue_data.csv',index=False)
```
